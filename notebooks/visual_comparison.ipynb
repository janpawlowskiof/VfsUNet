```python
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np

rows = 0
columns = 4

images = []
descriptions = []


def show_images():
    fig, axs = plt.subplots(rows, columns, figsize=(columns * 4, rows * 4))
    fig.tight_layout()
    axs = axs.flatten()

    for ax, im, desc in zip(axs, images, descriptions):
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.imshow(im, interpolation='nearest')
        # ax.set_title(desc, y=-0.1)

    fig.subplots_adjust(wspace=0.05, hspace=0.05)


# size_x, size_y = 128, 128
image_size = 128
```

```python
images, descriptions = [], []


def add_images(paths, start_x, start_y, desc):
    global rows
    rows += 1

    frames = [
        cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB) for path in paths
    ]

    full_frame = cv2.rectangle(
        frames[0].copy(),
        (start_x, start_y),
        (start_x + image_size, start_y + image_size),
        (255, 0, 0),
        6
    )
    # full_frame = cv2.resize(full_frame, (image_size * 16 // 9, image_size))

    images.extend(
        [full_frame] + [
            frame[start_y:start_y + image_size, start_x:start_x + image_size, :]
            for frame in frames
        ]
    )

    descriptions.extend(
        [
            desc, "Reference frame", "VFS-UNet",  "DBF + SAO", "No post-processing"
        ]
    )
```
