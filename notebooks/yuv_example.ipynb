```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

bgr = cv2.imread("/mnt/nfs_svtai10-nvme1n1p1/jpawlowski/gta_v/train/ai_hevc_qp=37_split_png/GrandTheftAutoV2021.11.07-11.27.52.10/000382.png")
yuv = cv2.cvtColor(bgr, cv2.COLOR_BGR2YUV)

y = np.copy(yuv)
y[:, :, 1] = 128
y[:, :, 2] = 128
y = cv2.cvtColor(y, cv2.COLOR_YUV2BGR)


u = np.copy(yuv)
u[:, :, 0] = 128
u[:, :, 2] = 128
u = cv2.cvtColor(u, cv2.COLOR_YUV2BGR)

v = np.copy(yuv)
v[:, :, 0] = 128
v[:, :, 1] = 128
v = cv2.cvtColor(v, cv2.COLOR_YUV2BGR)

cv2.imwrite("images/yuv/y_compressed.png", y)
cv2.imwrite("images/yuv/u_compressed.png", u)
cv2.imwrite("images/yuv/v_compressed.png", v)

plt.imshow(y)
plt.show()
plt.imshow(u)
plt.show()
plt.imshow(v)
plt.show()
```
